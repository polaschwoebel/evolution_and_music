```python
import os
import numpy as np
import pickle
from sklearn.decomposition import PCA
import itertools
import matplotlib.pyplot as plt
% matplotlib inline
from pylab import rcParams
rcParams['figure.figsize'] = 12, 8
```

```python
# read in all the files into a matrix (keep track of generations, could be just by indices)

# PCA on the matrix -> dim red to 2 

# plot with colours depending on generation
```

```python
i = 0
for filename in os.listdir('genotypes'):
    if filename.endswith('.p'):
        if i == 0:
            print(filename)
            all_gens = np.array(np.array(pickle.load( open( "genotypes/" + filename, "rb" ))))
            print(all_gens.shape)
        else:
            print(filename)
            individuals_here = np.array(pickle.load( open( "genotypes/" + filename, "rb" )))
            print(individuals_here.shape)
            all_gens = np.vstack([all_gens, individuals_here])
        i += 1
```

```python
pca = PCA(n_components=2)
reduced = pca.fit_transform(all_gens)
```

```python
color_words = ['k','r','g','m','c']
```

```python
colors = [[c]*5 for c in range(100)]
```

```python
colors = list(itertools.chain(*colors))
```

```python
colors
```

```python
x = reduced[:, 0].reshape((1, 500))
y = reduced[:, 1].reshape((1, 500))
```

```python
plt.scatter(x, y, c = np.array(colors), s=50)
plt.colorbar()
```

```python
plt.scatter(x, y, c = np.array(colors), s=50)
line1 = plt.Line2D(range(10), range(10), marker='o', color=colors[0])
line2 = plt.Line2D(range(10), range(10), marker='o',color=colors[10])
line3 = plt.Line2D(range(10), range(10), marker='o',color=colors[20])
line4 = plt.Line2D(range(10), range(10), marker='o',color=colors[30])
line5 = plt.Line2D(range(10), range(10), marker='o',color=colors[40])
plt.legend((line1,line2,line3, line4, line5),('color1','color2', 'color3', 'color4', 'color5'),numpoints=1, loc=1)
```

```python
plt.scatter(x, y, c = np.array(colors), s=50)
line1 = plt.Line2D(range(10), range(10), marker='o', color=color_words[0])
line2 = plt.Line2D(range(10), range(10), marker='o',color=color_words[1])
line3 = plt.Line2D(range(10), range(10), marker='o',color=color_words[2])
line4 = plt.Line2D(range(10), range(10), marker='o',color=color_words[3])
line5 = plt.Line2D(range(10), range(10), marker='o',color=color_words[4])
plt.legend((line1,line2,line3, line4, line5),('color1','color2', 'color3', 'color4', 'color5'),numpoints=1, loc=1)
```
