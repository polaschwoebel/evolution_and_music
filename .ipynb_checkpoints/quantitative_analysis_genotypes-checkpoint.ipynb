```python
import os
import numpy as np
import pickle
from sklearn.decomposition import PCA
import itertools
import matplotlib.pyplot as plt
% matplotlib inline
from pylab import rcParams
rcParams['figure.figsize'] = 12, 9
```

```python
# read in all the files into a matrix (keep track of generations, could be just by indices)

# PCA on the matrix -> dim red to 2 

# plot with colours depending on generation
```

```python
i = 0
for filename in os.listdir('genotypes'):
    if filename.endswith('.p'):
        if i == 0:
            print(filename)
            #all gens a list of index, generations
            all_gens = {filename[12:-2]:np.array(np.array(pickle.load( open( "genotypes/" + filename, "rb" ))))}
        else:
            print(filename)
            individuals_here = np.array(pickle.load( open( "genotypes/" + filename, "rb" )))
            all_gens[filename[12:-2]] = individuals_here
        i += 1
```

```python
print(all_gens)
```

```python
data = np.array(all_gens['1'])
for i in range(2, 101):
    data = np.vstack([data, all_gens[str(i)]])
```

```python
data.shape
```

```python
pca = PCA(n_components=2)
reduced = pca.fit_transform(data)
```

```python
color_words = ['k','r','g','m','c']
```

```python
colors = [[c]*5 for c in range(100)]
```

```python
colors = list(itertools.chain(*colors))
```

```python
colors
```

```python
len(colors)
```

```python
x = reduced[:, 0].reshape((1, 500))
y = reduced[:, 1].reshape((1, 500))
```

```python
fig, ax = plt.subplots()
plt.scatter(x, y, c = np.array(colors).reshape(1, 500), s=50)
plt.colorbar()
plt.grid(True)
#plt.xticks(np.array(['','', '', '', '','' ]))
#plt.yticks([])
#empty_string_labels =
#ax.set_xticklabels(empty_string_labels)



ax.tick_params(labelbottom='off', labelleft = 'off')    

plt.show()
```
